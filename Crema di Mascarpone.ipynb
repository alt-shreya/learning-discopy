## Import modules: `Ob, Ty, Box, Id, Drawing`

- `Ob`: (object) can be thought of as a basic ingredient in a recipe, it is defined only by its name.
- `Ty`: (type) is simply a sequence of objects, i.e. a bunch of ingredients side by side.
- `Box`: to draw a Box with some types as dom (domain) and cod (codomain) which represent the inputs and outputs of a process. 
- `Id`: (identity) for a type, i.e. just some parallel wires
- `Drawing`: used to draw the diagram

```python
from discopy import Ob, Ty, Box, Id, Drawing
```

# Diagram of the Recipe



### Define objects and types as per the diagram

```python
# define some types
yoghurt, applesauce = Ty('yoghurt'), Ty('applesauce')
sugar, mascarpone = Ty('sugar'), Ty('mascarpone')
yoghurty_paste, masca_mixture = Ty('yoghurty_paste'), Ty('masca_mixture')
crema_di_mascarpone = Ty('crema_di_mascarpone')
# not defined here: mixture, thick paste


# define the objects
assert sugar.objects == [Ob('sugar')]
assert yoghurt.objects == [Ob('yoghurt')]
assert applesauce.objects == [Ob('applesauce')]
assert mascarpone.objects == [Ob('mascarpone')]
```

### Define the reactants of the processes as types

```python
# define the combinations
assert sugar @ yoghurt == Ty(Ob('sugar'), Ob('yoghurt'))
assert applesauce @ mascarpone == Ty(Ob('applesauce'), Ob('mascarpone'))
assert yoghurty_paste @ masca_mixture == Ty(Ob('yoghurty_paste'), Ob('masca_mixture'))
```

### Describe the first process, beat, which involves sugar and yoghurt, and leads to the formation of a yoghurty paste

```python
beat = Box('beat', sugar @ yoghurt, yoghurty_paste)
beat.draw(figsize=(2, 2))
```

### Describe the second process, stir, which involves applesauce and mascarpone, and leads to the formation of a mascarpone mixture.

```python
stir = Box('stir', applesauce @ mascarpone, masca_mixture)
stir.draw(figsize=(2, 2))
```

### Describe the third process, fold, which involves yoghurty paste and the mixture of mascarpone, and leads to the formation of crema di mascarpone.

```python
fold = Box('fold', yoghurty_paste @ masca_mixture, crema_di_mascarpone)
fold.draw(figsize = (4,2))
```

## Describe a tensor mixture, using operator `@` which draws the processes side by side

```python
tensor_mixture = beat @ stir

drawing.equation(tensor_mixture, figsize = (8,2))
```

## Describe a sequence, using operator `>>` which draws the processes side by side

```python
sequence = tensor_mixture >> fold
drawing.equation(sequence, figsize = (10,4))
```
